## Visualization to compare the price changes between different  cities in Norway on 07 November 2022

```python
# To visualize the data, hourly rate of electricity price changing of each different cities in Norway. We can create a line of plot using 'plot()' method of the 'matplotlib' library. The plot displays the data in 'Oslo', 'Kr.sand',  Bergen, Molde, Tr.heim, and Tromsø' columns of the 'dataFrame_df_hours'. Each line corresponds to a different city and is plotted with a different color.

# Create a figure with one axis of the 'fig' figure. The 'plot()' method takes two arguments: the 'x' values and the 'y' values. In this case, the 'x' values are values in 'hour' variable, and the 'y' values are the values in the country columns of the dataFrame, 'dataFrame_df_hours'.
fig, ax = plt.subplots()

# The x-values of the plot are determined by the 'hour' variable, which is created by selecting the 'time' column from the 'dataFrame_df_hours' DataFrame.
hour = df_hours['time']

x = np.arange(len(hour))

# Create the bar chart
ax.plot(x, df_hours['Oslo'], label='Oslo', color='green')
ax.plot(x, df_hours['Kr.sand'], label='Kr.sand', color='red')
ax.plot(x, df_hours['Bergen'], label='Bergen', color='orange')
ax.plot(x, df_hours['Molde'], label='Molde', color='pink')
ax.plot(x, df_hours['Tr.heim'], label='Tr.heim', color='blue')
ax.plot(x, df_hours['Tromsø'], label='Tromsø', color='yellow')

# Set features like labels and titles are added in separate steps
ax.set_ylabel('Price(EU)')
ax.set_xlabel('Hour period (start at 00:00:00 - end at 24:00:00)')
ax.set_title('Hourly rate of electricity price changing of each different cities in Norway')

# The legend is added to the plot using the 'legend' method of the 'ax' object
ax.legend()

# Make the plot appear in the screen by using 'show()' method of the 'matplotlib.pyplot' module.
plt.show()

# If we take a closer look at each component of the prediction of electricity cost(hourly rate) of different cities in Norway on 17 November 2022, we can validate that 'Oslo', 'Kr.sand' and 'Bergen' have a match price's data, as the same with 'Molde' and 'Tr.heim' The trend of 'Oslo', 'Kr.sand' and 'Bergen' and  'Molde' and 'Trondheim' start at the similar figure between 36,37 and 28,87 while the electricity price in 'Trondheim' start at 17.67, constantly stable over the period between 17,67 - 18.04. During 8.00 am electricity price reach the highest peak at 85,86 in Oslo', 'Kr.sand' and 'Bergen', before falling after 10.00am, and declined dramatically to the less than the same figure as the started point, roughtly 4 Euro and end at the same figure as  'Molde' and 'Tr.heim' Molde' and 'Tr.heim' drop to the same figure as the elecity cost in 'Tromso' between  02-06.00 am then rise to roghtly 40 and stay constanly stable for a whole trend. In conclusion, the elecity reach the peak at 8 am in 'Oslo', 'Kr.sand' and 'Bergen' and  'Molde' and 'Tr.heim' , so it can siad that people consume a lot of eleccity in the early beginng morning of the day.
```

## Visualization to compare the price changes between different  cities in Norway, weekly.

```python
# In order to compare the weekly of price changing between different cities in Norway between November 2022, can create a bar plot using the 'bar()' method of the 'matplotlib' library. The plot displays the data in the 'Oslo', 'Kr.sand', 'Bergen', 'Molde', 'Tr.heim' and 'Tromsø'. Each bar corresponds to a different city and is plotted with a different color.

# The plot is created by calling the 'bar()' method on the 'ax' object,which is single axis of the 'fig' figure.
fig, ax = plt.subplots()

# The x-values of the bars is a weekly number of November 2022
week = df_weekly['weekly']
x = np.arange(len(week))

# DataFrame, and the width of the bars is set to '0.15'. The bar are plotted with an offset so that they are centered around the 'x' values.
width = 0.15

# Create the bar chart
ax.plot(x-2*width/2, df_weekly['Oslo'], width, label='Oslo', color='green')
ax.plot(x- width/2, df_weekly['Kr.sand'], width, label='Kr.sand', color='red')
ax.plot(x+ width/2, df_weekly['Bergen'], width, label='Bergen', color='orange')
ax.plot(x+2*width/2, df_weekly['Molde'], width, label='Molde', color='pink')
ax.plot(x-2*width/2, df_weekly['Tr.heim'], width, label='Tr.heim', color='blue')
ax.plot(x-width/2, df_weekly['Tromsø'], width, label='Tromsø', color='yellow')

# Set features like labels and titles are added in separate steps
ax.set_ylabel('Price(EU)')
ax.set_title('Weekly rate changes of electricity price of different city in Norway')

# This to ensure that we have one trick per week, otherwise we get fewer. The a-xis tick lables are set using the 'set_xticks()' set_xticklabels() methods of the ax object.
ax.set_xticks(x)
ax.set_xticklabels(week.astype(str).values, rotation='vertical')

# The legend is added to the plot using the 'legend()' method of the 'ax' object.
ax.legend()

# Finally, the plot is displayed using the 'show()' method of matplotlib.pyplot
plt.show()


# The chart highlight the weekly rate change of electricity price, different city in Norway between 'November 2022'. Information is divided into six groups of city by the prediction of electricity price. Overall, the prediction price of electricity in 'Oslo', 'Kr.sand' 'Bergen' and Oslo was predicted having similar a high electricity consumption over the period, dramatically growth one time higher from week 40 at roughly 90,00 to 170,00 on week 41 with small different gap of prices before gradually fall to the same figure at 140,00 on week 42, 120,00 at week 43 and around 70,00 accordingly. While the trend 'Mold', 'Tr.heim' and 'Tromso' began at around 20,00 then fluctuated decrease to under 20,00 before 'Molde' and 'Tr.heim' had a half higher price than Tromso throughout the week 42 and 43, fell slightly in final 3 weeks to approximately 30,00 and 'Tromso' slight fall to under 20,00
```
